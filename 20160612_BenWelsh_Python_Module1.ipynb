```python
import pandas as pd
```

```python
props = pd.read_csv("http://www.firstpythonnotebook.org/_static/committees.csv")
```

```python
contribs = pd.read_csv("http://www.firstpythonnotebook.org/_static/contributions.csv")
```

```python
prop = props[props.prop_name == 'PROPOSITION 064- MARIJUANA LEGALIZATION. INITIATIVE STATUTE.']
```

```python
merged = pd.merge(prop, contribs, on="calaccess_committee_id")
```

```python
support = merged[merged.committee_position == 'SUPPORT']
```

```python
oppose = merged[merged.committee_position == 'OPPOSE']
```

## Which committees have raised the most money?

To reformat data as a pretty DataFrame, use the reset_index method:

```python
merged.groupby("committee_name_x").amount.sum().reset_index().sort_values("amount", ascending=False)
```

```python
merged.groupby(["contributor_firstname", "contributor_lastname", "committee_position"]).amount.sum().reset_index().sort_values("amount", ascending=False)
```

## What percentage of donations came from people who live outside of California?

```python
out_of_state = merged[merged.contributor_state != 'CA']
```

```python
in_state = merged[merged.contributor_state == 'CA']
```

Check out top ten biggest out of state donors

```python
out_of_state.groupby(["contributor_firstname", "contributor_lastname", "committee_position"]).amount.sum().reset_index().sort_values("amount", ascending=False).head(10)
```

```python
'{:.2%}'.format(out_of_state.amount.sum()/merged.amount.sum())
```

## What are the top employers of donors who gave for and against the measure?

```python
support.groupby(["contributor_employer"]).amount.sum().reset_index().sort_values("amount", ascending=False).head(10)
```

```python
oppose.groupby(["contributor_employer"]).amount.sum().reset_index().sort_values("amount", ascending=False).head(10)
```

## Which committee had the fewest donors?

```python
merged.groupby(["committee_name_x", "calaccess_filing_id"]).contributor_lastname.agg('count').reset_index().sort_values("contributor_lastname").head(10)
```
